## Keypoint detection

### TODO: Detect and display the predicted keypoints

After each face has been appropriately converted into an input Tensor for your network to see as input, you'll wrap that Tensor in a Variable() and can apply your `net` to each face. The ouput should be the predicted the facial keypoints. These keypoints will need to be "un-normalized" for display, and you may find it helpful to write a helper function like `show_keypoints`.
```python
def show_all_keypoints(image, predicted_key_pts):
    """Show image with predicted keypoints"""
    predicted_key_pts = predicted_key_pts.data
    # un-transform the predicted key_pts data
    predicted_key_pts = predicted_key_pts.numpy()
    # undo normalization of keypoints  
    predicted_key_pts = predicted_key_pts*50.0+100
    
    # image is grayscale
    plt.imshow(np.squeeze(image), cmap='gray')
    plt.scatter(predicted_key_pts[:, 0], predicted_key_pts[:, 1], s=20, marker='.', c='m')
    
    plt.show()
```

```python
def rescale(image, size):
    h, w = image.shape[:2]
    if isinstance(size, int):
        if h > w:
            new_h, new_w = size * h / w, size
        else:
            new_h, new_w = size, size * w / h
    else:
        new_h, new_w = size

    new_h, new_w = int(new_h), int(new_w)

    img = cv2.resize(image, (new_w, new_h))
    
    return img
```

```python
def to_tensor(image):
    # if image has no grayscale color channel, add one
    if(len(image.shape) == 2):
        # add that third color dim
        image = image.reshape(image.shape[0], image.shape[1], 1)
            
    # swap color axis because
    # numpy image: H x W x C
    # torch image: C X H X W
    image = image.transpose((2, 0, 1))
    
    return torch.from_numpy(image)
```

```python
image_copy = np.copy(image)

# offset for the extension of roi 
offset = 40

# loop over the detected faces from your haar cascade
for (x,y,w,h) in faces:
    
    # Select the region of interest that is the face in the image 
    roi = image_copy[y-offset:y+h+offset, x-offset:x+w+offset]
    
    ## DONE: Convert the face region from RGB to grayscale
    roi_copy = cv2.cvtColor(roi, cv2.COLOR_RGB2GRAY)

    ## DONE: Normalize the grayscale image so that its color range falls in [0,1] instead of [0,255]
    roi_copy =  roi_copy/255.0
    
    ## DONE: Rescale the detected face to be the expected square size for your CNN (224x224, suggested)
    roi_copy = rescale(roi_copy, 224)
    
    ## DONE: Reshape the numpy image shape (H x W x C) into a torch image shape (C x H x W)
    roi_copy = to_tensor(roi_copy)
    roi_copy = roi_copy.unsqueeze(0).type(torch.FloatTensor)
    
    ## DONE: Make facial keypoint predictions using your loaded, trained network 
    ## perform a forward pass to get the predicted facial keypoints
    predicted_keypoints = net(roi_copy)
    
    # reshape the predicted keypoints
    predicted_keypoints = predicted_keypoints.view(68, -1)
    
    ## DONE: Display each detected face and the corresponding keypoints        
    show_all_keypoints(roi_copy, predicted_keypoints)    
```
